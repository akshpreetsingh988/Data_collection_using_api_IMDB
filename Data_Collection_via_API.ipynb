```python
List_of_similar_movies = []
for i in new_df['id']:
  link = 'https://api.themoviedb.org/3/movie/'+ str(i) +'/similar?api_key=' + api_key
  similar_movie_name = requests.get(link).json()['results']
  similar_movie_id = []

  for j in range(len(similar_movie_name)):
    similar_movie_id.append(similar_movie_name[j]['id'])
    if j == 4:
      break

  for k in similar_movie_id:
    List_of_similar_movies.append([i,k])    
```

```python
List_of_similar_movies
```

```python
len(List_of_similar_movies)
```

```python
val = 0
while(val < len(List_of_similar_movies)):
  similar_mov = List_of_similar_movies[val]

  if similar_mov[::-1] not in List_of_similar_movies:
    val = val +1
    continue
  if similar_mov[0] < similar_mov[1] : 
    List_of_similar_movies.remove(similar_mov[::-1])
    val = val + 1
  else:
    List_of_similar_movies.remove(similar_mov) 
       
```
